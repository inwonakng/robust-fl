```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
to_record = ['random_sample_simple_' + str(i) for i in range(1, 8)]
to_record.append('random_sample_simple_0_5')
name_remapping = dict(*(('random_sample_simple_' + str(i), 'Simple Random Sample (p = 0.' + str(i) + ')') for i in range(8)), ('random_sample_simple_0_5', 'Simple Random Sample (p = 0.05)'))

result = pd.concat([
    pd.read_csv(f'../output/{name}/mnist_0_1/no_poison_no_straggle/report.csv').assign(**{
        "Aggregator Type": name_remapping[name]
    }) for name in to_record
]).rename(columns = dict(accuracy_score="Accuracy Score", round="Round"))
```

```python
sns.relplot(data=result, x="Round", y="Accuracy Score", kind="line", hue="Aggregator Type", height = 3, aspect= 1)
```
