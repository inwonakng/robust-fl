```python
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.decomposition import PCA
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
points = np.random.random((100, 10000))
```

```python
kmeans_clusters = KMeans(n_clusters=5).fit_predict(points)
agg_clusters = AgglomerativeClustering().fit_predict(points)
```

```python
points_2d = PCA(n_components=2).fit_transform(points)
```

```python
sns.scatterplot(
    x = points_2d[:,0],
    y = points_2d[:,0],
    hue = kmeans_clusters
)
```

```python
sns.scatterplot(
    x = points_2d[:,0],
    y = points_2d[:,0],
    hue = agg_clusters
)
```
